# PCA - Fashion MNIST

En éste ejercicio vamos a aplicar PCA y K-means al digits dataset similar a MNIST, que incluye imágenes de dígitos de 8x8 píxeles. La idea es intentar recopilar información sobre la distribución de las imágenes, usando visualizaciones que permitan clarificar cómo interpretar los resultados obtenidos por PCA.

6. Graficar los clusters de dígitos en 2 y 3 dimensiones usando los componentes obtenidos en PCA.

```python
# (1) Estandarizamos
X_norm = normalize(X)
# (1) Aplicamos PCA
PCAModel = PCA(n_components=0.9,svd_solver='full')
PCAModel.fit(X_norm)
# (1) Cantidad de componentes para contemplar el 90% de varianza 
m = PCAModel.components_.shape[0]
print("Cantidad de componentes para explicar el 90% de varianza: " + str(m))
```

```python
# (2) Scree plot
plt.plot(range(m), PCAModel.explained_variance_ratio_[:m])
plt.plot(range(m), np.cumsum(PCAModel.explained_variance_ratio_[:m]))
plt.title("Varianza explicada por componente y Cumulativa ", fontsize=10)
```

```python
# (3) Visualización de los primeros 5 componentes (TODO: explicación)
plt.figure(figsize=(10,10),dpi=100)
for i in range(5):
    comp = PCAModel.components_[i,:].ravel().reshape((8, 8))
    plt.subplot(1,5,i+1)
    plt.imshow(comp,cmap='gray')
    plt.axis('off')
    titleStr = str(100*PCAModel.explained_variance_ratio_[i])[0:5]
    plt.title(titleStr + '% explained')
```

```python
# (4) Original vs reconstruida
plt.figure(figsize=(10,4),dpi=100)
for i in range(5):
    digit = X[i,:].reshape((8, 8))
    plt.subplot(2,5,i+1)
    plt.imshow(digit,cmap='gray')
    plt.title('Original ' + str(i)) 
    plt.axis('off')
  
plt.figure(figsize=(10,4),dpi=100)
for i in range(5):
    digit = X[i,:]
    digit_t = PCAModel.transform(digit.reshape(1, -1))
    digit_r = PCAModel.inverse_transform(digit_t).reshape((8, 8))
    plt.subplot(2,5,i+1)
    plt.imshow(digit_r,cmap='gray')
    plt.title('Reconstruida ' + str(i)) 
    plt.axis('off')
    
plt.figure(figsize=(10,4),dpi=100)
for i in range(5,10):
    digit = X[i,:].reshape((8, 8))
    plt.subplot(2,5,i+1)
    plt.imshow(digit,cmap='gray')
    plt.title('Original ' + str(i)) 
    plt.axis('off')
  
plt.figure(figsize=(10,4),dpi=100)
for i in range(5,10):
    digit = X[i,:]
    digit_t = PCAModel.transform(digit.reshape(1, -1))
    digit_r = PCAModel.inverse_transform(digit_t).reshape((8, 8))
    plt.subplot(2,5,i+1)
    plt.imshow(digit_r,cmap='gray')
    plt.title('Reconstruida ' + str(i)) 
    plt.axis('off')
```

```python
# (5) matriz de correlación del dataset reducido
plt.figure(figsize=(10,4),dpi=100)
print(PCAModel.get_covariance().shape)
plt.imshow(PCAModel.get_covariance())
```

```python
# (6) Clusters de dígitos en 2 y 3 dimensiones

%matplotlib

X_pca = PCAModel.transform(X_norm)
plt.figure(figsize=(12,10))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolor='none', alpha=0.7, s=40, cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection='3d')
p = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, edgecolor='none', alpha=0.7, s=40, cmap=plt.cm.get_cmap('nipy_spectral', 10))
fig.colorbar(p)
```

```python
# (7) K-means para clusterizar los dígitos
from sklearn.cluster import KMeans
RANDOM_STATE = 17
n_clusters=10
kmeans = KMeans(n_clusters=n_clusters, n_init=100, random_state=RANDOM_STATE)
kmeans.fit(X_pca)
cluster_labels = kmeans.labels_

# Qué pasa si hago trampa y paso como centroides iniciales un promedio de los clusters que conozco ?
```

```python
# (7) Graficar los resultados de K-means
#%matplotlib

X_pca = PCAModel.transform(X_norm)
plt.figure(figsize=(12,10))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, edgecolor='none', alpha=0.7, s=40, cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection='3d')
p = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=cluster_labels, edgecolor='none', alpha=0.7, s=40, cmap=plt.cm.get_cmap('nipy_spectral', 10))
fig.colorbar(p)
```

```python
# (7) Análisis de los resultados de K-means
# ** Usar indicadores vistos en el TP1 para medir el desempeño ?

from scipy import stats as st

global_acc = 0

clus2num = np.zeros(n_clusters,dtype=np.int32)

for i in range(n_clusters):
    group = y[cluster_labels == i]
    moda, count = st.mode(group)
    acc = np.round(count[0]*100/group.size,decimals=0)
    global_acc += acc
    print( "Cluster mode: " + str(moda[0]) + ". Accuracy: " + str(acc) + "%")
    clus2num[i] = moda[0]
    
global_acc /= n_clusters
print("Global accuracy: " + str(global_acc) + "%")
```

```python
# (8) Gráfico de inercia para obtener el número óptimo de clusters k
# Vemos la inercia (suma de las distancias de los miembros de un cluster a su centroide)

num = 21

inertia = []
for k in range(1, num):
    kmeans = KMeans(n_clusters=k, n_init=100, random_state=RANDOM_STATE).fit(X_pca)
    inertia.append(np.sqrt(kmeans.inertia_))
    
plt.plot(range(1, num), inertia, marker='s')
```

```python
# (9) Analizar visualmente los límites del cluster de algún dígito y "generar" 
# artificialmente el dígito dándole valores a los primeros dos componentes de PCA

# comp = PCAModel.components_
# n9_r = (-0.3)*comp[0, :] + (-0.2)*comp[1, :] # ? Seria hacer lo mismo aća ? Porque el resultado es diferente ?

x = np.zeros(21)
x[:2] = [-0.3, -0.2]
n9_r = PCAModel.inverse_transform(x)

x[:2] = [0.3, -0.25]
n6_r = PCAModel.inverse_transform(x)

x[:2] = [0, -0.4]
n0_r = PCAModel.inverse_transform(x)

fig = plt.figure(figsize=(12,12))

plt.subplot(1,3,1)
plt.imshow(n9_r.reshape(8, 8),cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(n6_r.reshape(8, 8),cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(n0_r.reshape(8, 8),cmap='gray')
plt.axis('off')
```

```python
# Extra: sacar un numero aleatorio del dataset y detectar que numero es.
%matplotlib inline

fig = plt.figure(figsize=(4,4))

i = np.random.randint(X.shape[0])
print("Random dataset digit: " + str(y[i]))
plt.imshow(X[i, :].reshape((8, 8)), cmap='gray')
plt.show()

i_norm = normalize(X[i, :].reshape(1, -1))
i_t = PCAModel.transform(i_norm)
pred = kmeans.predict(i_t)
number = clus2num[pred]

print("Detected digit: " + str(number))
```
